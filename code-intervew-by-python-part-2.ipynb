```python
def getDataFromUser():
    D = {}
    while True:
        studentId = input('Enter sutdent ID:')
        markslist = input('Enter the marks by, comma separted values: ')
        moreStudents = input('Enter "no" to  quit insertion:')
        if studentId in D:
            print(studentId , 'is alreday insterted')
        else:
            D[studentId] = markslist.split(',')
        if moreStudents.lower() == 'no':
            return D
            
        
        
```

```python
sudentData
```

```python
def getAvgMarks(D):
    avgMarks = {}
    for x in D:
        L  = D[x]
        s = 0
        for marks in L:
            s += int(marks)
        avgMarks[x]= s/len(L)
        
    return avgMarks
            
```

```python
def calculate_averages(student_data):
    results = {}
    for name , grads in student_data.items():
        total_sum = sum(int(g) for g in grads)
        average  =  total_sum / len(grads)
        results[name] = average
    return results
```

```python
agvM = calculate_averages(sudentData)
```
